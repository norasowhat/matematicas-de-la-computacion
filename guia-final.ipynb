```python
def sistemaVerificacionPatrones():
    secuencia = input("ingresa la cadena (a's y b's): ")
    modo_operacion = 100  

    print("analizando patrón de entrada...")
   
    if '"' in secuencia:
        print("patrón con formato incorrecto")
        return 
    
    for caracter in secuencia:
        if caracter not in {"a","b"}:
            print("elemento no reconocido en el patrón")
            return 
        
        match modo_operacion:
            case 100:  
                if caracter == "a": modo_operacion = 101
                if caracter == "b": modo_operacion = 102

            case 101:  
                if caracter == "a": modo_operacion = 100
                if caracter == "b": modo_operacion = 103

            case 102:   
                if caracter == "a": modo_operacion = 103
                if caracter == "b": modo_operacion = 100

            case 103:   
                if caracter == "a": modo_operacion = 102
                if caracter == "b": modo_operacion = 101

        print("procesando elemento", caracter, "Modo:", modo_operacion)

    if modo_operacion == 102:
        print("patrón validado: comportamiento normal")
    else:
        print("patrón rechazado: anomalía detectada")
```

```python
if __name__ == "__main__":
    sistemaVerificacionPatrones()
```
