identificación de números primos

```python
def es_primo(num):
    if num <= 1:
        return False
    if num <= 3:
        return True
    if num % 2 == 0 or num % 3 == 0:
        return False
    
    i = 5
    while i * i <= num:
        if num % i == 0 or num % (i + 2) == 0:
            return False
        i += 6
    
    return True
```

```python
num = int(input("dame un número: "))

if es_primo(num):
    print(f"{num} es primo")
else:
    print(f"{num} no es primo")
```

teorema de legendre

```python
def contar_primos(n):
    es_primo = [False] * (n + 1)

    for i in range(2, n + 1):
        es_primo[i] = True

    p = 2
    while p * p <= n: 
        if es_primo[p]:
            for i in range(p * p, n + 1, p):
                es_primo[i] = False
        p += 1

    contador = 0
    for i in range(2, n + 1):
        if es_primo[i]:
            contador += 1

    return contador
```

```python
n = int(input("dame un número: "))
resultado = contar_primos(n)
print(f"{n} tiene un total de {resultado} números primos")
```
