```python
def signo(num):
    if num > 0:
        return 1
    elif num < 0:
        return -1
    else:
        return 0
```

```python
def euclides_extendido(a, b):
    c = abs(a)
    d = abs(b)

    c1 = 1
    c2 = 0

    d1 = 0
    d2 = 1

    while d != 0:
        q = c // d
        r = c - q * d
        r1 = c1 - q * d1
        r2 = c2 - q * d2
        
        c, d = d, r
        c1, d1 = d1, r1
        c2, d2 = d2, r2
    
    mcd=abs(c)

    s=c1/signo(a)*signo(c)

    t=c2/signo(b)*signo(c)

    return int(mcd),int(s),int(t)

```

```python
a=int(input("dame el valor para a"))
b=int(input("dame el valor para b"))

mcd, s, t = euclides_extendido(a, b)
```
